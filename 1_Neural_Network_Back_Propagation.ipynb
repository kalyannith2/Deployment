```python
# Create your first MLP in Keras
from keras.models import Sequential   #create dummy network
from keras.layers import Dense
import numpy


```

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# load pima indians dataset
dataset = numpy.loadtxt("/content/pima-indians-diabetes.data.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
dataset
```

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy','precision'])
```

```python
# Fit the model
model.fit(X, Y, validation_split=0.33, epochs=500, batch_size=20)
```

```python
# evaluate the model
scores = model.evaluate(X, Y)

```

```python
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
