# chi-square test

```python
satisfaction=['very satisfied','satisfied','neutral','unsatisfied','very unsatisfied']
smart_thermostat=[50,80,60,30,20]
smart_light=[70,100,90,50,50]
total=[120,180,150,80,70]
```

```python
import pandas as pd
data=pd.DataFrame({'satisfaction':satisfaction,
                   'smart_thermostat':smart_thermostat,
                   'smart_light':smart_light,
                   'total':total})
```

```python
data
```

```python
crosstable =pd.crosstab(data['smart_thermostat'],data['smart_light'])
crosstable
```

```python
#chi_square test
from scipy.stats import chi2_contingency
chi,p,dof,exp_freq = chi2_contingency(crosstable)
print(chi,p,exp_freq)
```

```python
#hypothesis decision
if p<0.05:
    print('Reject H0: there is a significant association')
else:
    print('Accept H0: no significant association')     
```

# hypotheses test

```python
# given data
s_mean=3050
mu=600
cost_model=1000+(5*600)
sigma=5*25
n=25 #sample size
alpha=0.05
```

```python
#H0: mu = cost_model
#H1: mu =! cost_model
```

```python
#calculate the test statistic
import numpy as np
z = (s_mean - cost_model)/(sigma/np.sqrt(n))
z
```

```python
# critical value (z)
from scipy.stats import norm
z_crit = norm.ppf(1-alpha/2)
z_crit
```

```python
# decision
if abs(z)>z_crit:
    decision = 'reject the null hypotesis'

else:
    decision ='accept the null hypotesis'

decision
```
