```python
def recommend_movies(user_id, user_sim_df, movies_df, num_recommendations=5):


    # Find the most similar user to the target user
    similar_users = user_sim_df.loc[user_id].sort_values(ascending=False).index[0:]  # Exclude the user itself

    # Movies rated by the target user
    movies_watched_by_user = movies_df[movies_df['userId'] == user_id]['movie'].tolist()

    # Movies watched by similar users but not by the target user
    recommendations = []
    for similar_user in similar_users:
        movies_watched_by_similar_user = movies_df[movies_df['userId'] == similar_user]['movie'].tolist()
        for movie in movies_watched_by_similar_user:
            if movie not in movies_watched_by_user and movie not in recommendations:
                recommendations.append(movie)
                if len(recommendations) == num_recommendations:
                    return recommendations
    return recommendations

```

```python
user_id_to_recommend = int(input("enter user id "))
recommendations = recommend_movies(user_id_to_recommend, user_sim_df, movies_df,4)
print(f"Recommended movies for user {user_id_to_recommend}: {recommendations}")
```
