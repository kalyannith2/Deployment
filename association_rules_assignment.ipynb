```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df=pd.read_excel('Online retail.xlsx',sheet_name='Sheet1')
df.to_csv('Online retail.csv')
df.head()
```

```python
# Convert the dataframe into a list of transactions
transactions = df.values.tolist()

# Flatten the list of transactions and get unique items
all_items = [item for sublist in transactions for item in sublist[0].split(',')]
unique_items = sorted(list(set(all_items)))

# Create a one-hot encoded dataframe
oh_df = pd.DataFrame(0, index=range(len(transactions)), columns=unique_items)

for i, transaction in enumerate(transactions):
    for item in transaction[0].split(','):
        oh_df.at[i, item] = 1

display(oh_df.head())
```

```python
fre_item=apriori(oh_df,min_support=0.05,use_colnames=True)
fre_item
```

```python
rules=association_rules(fre_item,metric='lift',min_threshold=1)
rules
```

```python
rules.sort_values('lift',ascending=False)[0:20]
```

```python
1.	What is lift and why is it important in Association rules?
A.  Lift is a matric that measures how much more likely two items occur together than would be expected if they were independent.
    It tells us the strength of the relationship between the items in a rule.

    Importance:
    1.Helps identify strong and meaningful associations in large datasets.
    2.Used to rank rules and filter out weak or spurious associations.
    3.More informative than just support or confidence because it accounts for item independence.
```

```python
2.	What is support and Confidence. How do you calculate them?
A.  Support measures how frequently an itemset appears in the dataset.
    It tells us how popular or common an item or combination of items is.
    Support(Aâ†’B)=number of transactions containing both A and B / total number of transactions

     Confidence measures how often items in B appear in transactions that contain A.
     It tells us the strength of the implication.
     Confidence(Aâ†’B)=number of transactions containing both A and B / number of transactions containing A
```

```python
3.	What are some limitations or challenges of Association rules mining?
A.  Association rule mining is useful for discovering patterns, but itstruggles with large datasets,
    rare items, setting thresholds, computational cost, and it shows correlation rather than causation.
```
