```python
def recommend_anime(anime_id, user_sim_df, df, num_recommendations=5):


    # Find the most similar user to the target user
    similar_users = user_sim_df.loc[anime_id].sort_values(ascending=False).index[0:]  # Exclude the user itself

    # anime rated by the target user
    anime_watched_by_user = df[df['anime_id'] == anime_id]['name'].tolist()

    # anime watched by similar users but not by the target user
    recommendations = []
    for similar_user in similar_users:
        anime_watched_by_similar_user = df[df['anime_id'] == similar_user]['name'].tolist()
        for anime in anime_watched_by_similar_user:
            if anime not in anime_watched_by_user and anime not in recommendations:
                recommendations.append(anime)
                if len(recommendations) == num_recommendations:
                    return recommendations
    return recommendations

```

A. User-based collaborative filtering:- finds similar users and recommends their items.
   
  Item-based collaborative filtering:- finds similar items to those a user liked and recommends them.

2. What is collaborative filtering, and how does it work?

A. Collaborative filtering is a recommendation technique that makes prediction about a user`s interests based on the preferences or behaviors of similar users.
   It works by using a user - item interaction matrix, which store information such as ratings, clicks, or purchases.
   The system analyzes this matrix to find patterns of similarity - either between users or between items - and then uses those patterns to generate recommendations.

```python
# ===========================================
# Anime Recommendation System Evaluation
# ===========================================

# Import required libraries
import pandas as pd
from surprise import SVD, Dataset, Reader
from surprise.model_selection import train_test_split, cross_validate
from surprise import accuracy
from collections import defaultdict
import numpy as np
# Example: Anime dataset with columns ['anime_id', 'name','rating']
df = pd.read_csv('anime.csv')  # <-- Replace with your dataset file
print(df.head())

print(df.isnull().sum())
df = df.dropna(subset=['anime_id','name', 'rating'])


reader = Reader(rating_scale=(1, 10))  # ratings typically range 1–10
data = Dataset.load_from_df(df[['anime_id','name', 'rating']], reader)


# Train-Test Split

trainset, testset = train_test_split(data, test_size=0.2, random_state=42)


# Build Collaborative Filtering Model (SVD)
model = SVD(random_state=42)
model.fit(trainset)
predictions = model.test(testset)


# Evaluation Metrics
print("\n Model Evaluation:")
rmse = accuracy.rmse(predictions)
mae = accuracy.mae(predictions)

# Cross-validation for robust evaluation
print("\nCross-validation results:")
cross_val_results = cross_validate(model, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)


#  Precision@K and Recall@K

def precision_recall_at_k(predictions, k=10, threshold=7.5):
    """Return precision and recall at k metrics for each user."""
    # First map predictions to each user.
    user_est_true = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        user_est_true[uid].append((est, true_r))

    precisions = dict()
    recalls = dict()
    for uid, user_ratings in user_est_true.items():
        # Sort user ratings by estimated value
        user_ratings.sort(key=lambda x: x[0], reverse=True)
        # Count relevant items
        n_rel = sum((true_r >= threshold) for (_, true_r) in user_ratings)
        # Count recommended items in top k
        n_rec_k = sum((est >= threshold) for (est, _) in user_ratings[:k])
        # Count relevant and recommended
        n_rel_and_rec_k = sum(((true_r >= threshold) and (est >= threshold))
                              for (est, true_r) in user_ratings[:k])
        # Precision@K: Proportion of recommended items that are relevant
        precisions[uid] = n_rel_and_rec_k / n_rec_k if n_rec_k != 0 else 0
        # Recall@K: Proportion of relevant items that are recommended
        recalls[uid] = n_rel_and_rec_k / n_rel if n_rel != 0 else 0

    # Average over all users
    precision = np.mean(list(precisions.values()))
    recall = np.mean(list(recalls.values()))
    return precision, recall

precision, recall = precision_recall_at_k(predictions, k=10)
print(f"\n Precision@10: {precision:.4f}")
print(f" Recall@10: {recall:.4f}")


#  Generate Top-N Recommendations

def get_top_n(predictions, n=10):
    """Return top-N recommendation for each user."""
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))
    # Sort the predictions for each user and retrieve top n
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]
    return top_n

top_n = get_top_n(predictions, n=5)

# Example: show top 5 recommendations for first few users
for uid, user_ratings in list(top_n.items())[:5]:
    print(f"\nRecommended animes for user {uid}:")
    for (iid, est_rating) in user_ratings:
        print(f"  Anime {iid} with predicted rating {est_rating:.2f}")


#  Summary of Evaluation

print("\n📊 Evaluation Summary:")
print(f"RMSE: {rmse:.4f}")
print(f"MAE:  {mae:.4f}")
print(f"Precision@10: {precision:.4f}")
print(f"Recall@10: {recall:.4f}")
```
