```python
data=pd.read_csv("adult_with_headers.csv")
data.head()
```

```python
data=pd.get_dummies(data).astype('int')
data
```

```python
#standard scaling
from sklearn.preprocessing import StandardScaler
```

```python
scaler1=StandardScaler()
scaled_data=scaler1.fit_transform(data)
scaled_data
```

```python
#min_max scaling
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler2=MinMaxScaler()
scaled_data1=scaler2.fit_transform(data)
scaled_data1
```

```python
#isolation forest
from sklearn.ensemble import IsolationForest
```

```python
clf=IsolationForest(random_state=10,contamination=.01)
clf.fit(data)
```

```python
y_pred_outliers=clf.predict(data)
y_pred_outliers
```

```python
data['score']=clf.decision_function(data)
data['anomaly']=clf.predict(data)
```
