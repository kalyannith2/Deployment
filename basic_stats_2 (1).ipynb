```python
import numpy as np
import scipy.stats as st
```

```python
#given data
data=[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

# 99% confidence - sample standard deviation

```python
n=len(data)  #sample size
mean=round(np.mean(data),3)  #sample mean
```

```python
#significance level
alpha=0.01  #0.01 for 99% confidence
 #sample standard deviation
s_std=round(np.std(data,ddof=1),3)  
s_std
```

```python
#t-critical value for 99% confidence interval
t_crit=st.t.ppf(1-(alpha/2),n-1)
#margin of error
margin_err = t_crit * (s_std/np.sqrt(n))
#calculate confidence interval(lower bound, upper bound)
ci_t= (mean - margin_err, mean + margin_err)
ci_t
```

# 99% confidence - population standard deviation

```python
#significance level
alpha = 0.01
p_std=0.2  #population standard deviation
#z-critical value for 99% confidence interval
z_crit= st.norm.ppf(1-(alpha/2))
#margin of error
mar_err= z_crit * (p_std/np.sqrt(n))
ci_z= (mean-mar_err,mean+mar_err)
ci_z
```
