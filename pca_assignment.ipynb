```python
import pandas as pd
import numpy as np                                #PCA used to reduce the dimensions
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
df=pd.read_csv('wine.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df_std=scale(df)
df_std
```

```python
pca=PCA()
pca_values=pca.fit_transform(df_std)
pca_values
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
#scree plot
plt.plot(var1,color='red')
plt.show()
```

```python
pca.components_
```

```python
finaldf=pd.concat([df['Type'],pd.DataFrame(pca_values[:,0:2],columns=['PC1','PC2'])],axis=1 )
finaldf
```

```python
import seaborn as sns
sns.scatterplot(data=finaldf,x='PC1',y='PC2',hue='Type')
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

```

```python
# Apply PCA with optimal components
pca = PCA(n_components=3)
X_pca = pca.fit_transform(df)

kmeans_pca = KMeans(n_clusters=3, random_state=42)
labels_pca = kmeans_pca.fit_predict(X_pca)

# Evaluation
sil_score_pca = silhouette_score(X_pca, labels_pca)
db_score_pca = davies_bouldin_score(X_pca, labels_pca)
print("Silhouette PCA:", sil_score_pca, "DB Index PCA:", db_score_pca)
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels_pca, cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Clusters on PCA-transformed Data')
plt.show()
```
