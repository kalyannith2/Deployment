```python
#imputing
data1.loc[:,'LB'] =data1['LB'].fillna(data1.LB.median())
data1.loc[:,'AC'] =data1['AC'].fillna(data1.AC.median())
data1.loc[:,'DS'] =data1['DS'].fillna(data1.DS.median())
data1.loc[:,'DP'] =data1['DP'].fillna(data1.DP.median())
data1.loc[:,'MLTV'] =data1['MLTV'].fillna(data1.MLTV.median())
data1.loc[:,'NSP'] =data1['NSP'].fillna(data1.NSP.median())
data1.loc[:,'Width'] =data1['Width'].fillna(data1.Width.median())
data1.loc[:,'Tendency'] =data1['Tendency'].fillna(data1.Tendency.median())
```

```python
#IQR
q1=data1.AC.quantile(0.25)
q3=data1.AC.quantile(0.75)
IQR=q3-q1
ue=q3+1.5*IQR
le=q1-1.5*IQR
data1.loc[data1.AC>ue,'AC']=ue
data1.loc[data1.AC<le,'AC']=le
plt.boxplot(data1.AC)
```
