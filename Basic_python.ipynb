# prime number

```python
#take input from user
n=int(input('Enter a number: '))

#0 and 1 are not a prime numbers
if n==0 or n==1:
      print(n,'is not a prime number')
else:
      is_prime = True    #assume number is prime
      for i in range(2,n):
          if n%i == 0:
              is_prime = False  # then it's a prime
              break             #no need to checkfurther
      if is_prime:
          print(n,'is a prime number')
      else:
          print(n,'is a not a prime number')
    
```

# product of random numbers

```python
import numpy as np
import random

def product(a,b,y):
    if a*b == y:
       print('you answer is correct')
    else:
       print(' you are wrong, answer is' ,a*b) 
        
#generate two random numbers        
x=np.random.randint(1,20,2)
a,b=x

print('a=',a,'b=',b)

#ask the user for the product
y=int(input('what is the product of two numbers: '))

#now call the function
product(a,b,y)
```

# square of odd/even numbes

```python
#take input from user
n=int(input('enter a number between 100 to 200: '))

for i in range (100,201):
   if n%2 == 0:   #input number is even
      if i%2==0:   #check current number is even
         print(f'Even:{i}^2 = {i*i}')
      else:
         continue    #skip odd numbers

   else:
      if i%2!=0:   #check current number is odd
         print(f'Odd:{i}^2 = {i*i}')
      else:
         continue     #skip even numbers
      
```

# word counter

```python
from collections import Counter 

#take input from user
text = input('enter a sentence: ')
#converting text to lowercase to avoid case issues
text=text.lower()
#splitting sentence into list of words
words=text.split()
#countting each word
word_count=Counter(words)
#converting counter object to dictionary
count=dict(word_count)
print(count)
```

# palindrome


```python
def is_palindrome(text):
    #convert to lowercase and remove spaces
    text=str(text).lower().replace('','')
    return text==text[::-1]   #check if text is equal to its reverse

#take input from user
input=input('input: ')
print(is_palindrome(input))
```
